```python
# Function to clean 'review_point' and 'review_date'
def clean_data(df):
    df = df.reset_index()
    # Cleaning 'review_point' by extracting numeric value
    df['review_point'] = df['review_point'].apply(lambda x: int(re.findall(r'\d+', x)[-1]))
    
    # Cleaning 'review_date' by removing '작성일자'
    df['review_date'] = df['review_date'].str.replace('작성일자 ', '')
    
    # Check if '매장' is present in review_date and update purchase_place
    df['purchase_place'] = df.apply(lambda row: '매장' if '매장' in row['review_date'] else row['purchase_place'], axis=1)
    df['purchase_place'] = df.apply(lambda row: '선물' if '선물' in row['review_date'] else row['purchase_place'], axis=1)
    df['purchase_place'] = df.apply(lambda row: '체험단' if '체험단' in row['review_date'] else row['purchase_place'], axis=1)
    df['purchase_place'] = df['purchase_place'].fillna('온라인')

    #전체 \n제거
    df = df.replace(r'\n', '', regex=True)
    # Remove '매장' from review_date
    duplicate_mask = df.duplicated(subset=['user_name', 'product_name'], keep=False)
    
    # Set 'repurchase' to 1 for all duplicates
    df.loc[duplicate_mask, 'repurchase'] = 1
    df['repurchase'] = df['repurchase'].fillna(0)
    return df

# Applying the cleaning function to the DataFrame
cleaned_df = clean_data(total_review)
cleaned_df
```
