```python
import pandas as pd
from surprise.model_selection import train_test_split
from surprise import SVD
from surprise import accuracy
from surprise.model_selection import cross_validate
```

```python
#file_path = "C:\\Users\\82103\\Upstage AI Lab\\Olive_RecSys\\upstage-ailab-ml-project\\RecSys_Model\\temp_data\\konlpy.csv"
file_path = "/Users/soom/Workspace/Upstageailab/[ML]Olive_RecSys/upstage-ailab-ml-project/RecSys_Model/temp_data/sentiment_df.csv"
df = pd.read_csv(file_path, encoding="UTF-8")
df.drop("Unnamed: 0",axis=1, inplace= True)
df
```

```python
df['senti_score'] = df['Sentiment Score']+df['review_point'].fillna(0)
df
```

```python
from surprise import Dataset
from surprise import Reader

# rating_scale=(최소 평점, 최대 평점)을 설정해줍니다 (예시: 평점 범위가 1~5)
reader = Reader(rating_scale=(0.5, 5))

# userId, movieId, rating 컬럼만 선택하여 Surprise 데이터셋으로 변환
data = Dataset.load_from_df(df[['user_name', 'product_name', 'senti_score']], reader)

# Surprise 데이터셋을 학습에 사용하기 위한 trainset으로 변환
trainset = data.build_full_trainset()
```

```python
from surprise import SVD
from surprise import accuracy
from surprise.model_selection import train_test_split
import mlflow
import mlflow.sklearn
from itertools import product

mlflow.autolog()

# Set MLflow tracking URI and experiment
mlflow.set_tracking_uri("http://127.0.0.1:5000")
mlflow.set_experiment('Olive-RecSys')
# 학습 데이터와 테스트 데이터로 분할
trainset, testset = train_test_split(data, test_size=0.25)

n_factors = [5,10,15,20]
n_epochs = [1,3,5]

def new_func(a, l1, model):
    model.fit(trainset)

    # 테스트 데이터에 대한 예측
    predictions = model.test(testset)

    # 예측 성능 평가 (RMSE 계산)
    rmse = accuracy.rmse(predictions)

        # Log parameters and metrics
    mlflow.log_param("n_factors", a)
    mlflow.log_param("n_epochs", l1)
    mlflow.log_metric("rmse", rmse)
    return rmse

for a, l1 in product(n_factors, n_epochs):
    with mlflow.start_run(nested=True):
            # SVD 모델 생성 및 학습
        model = SVD(n_factors = a,n_epochs=l1)
        
        rmse =  new_func(a, l1, model)

        print(f"RMSE: {rmse}")
```

```python
# 새로운 사용자에 대해 영화 추천 (사용자의 평점을 예측하지 않은 영화 목록 중)
def get_top_n_recommendations(predictions, n=10):
    # 예측한 평점들을 정렬하여 상위 n개의 영화를 반환
    top_n = {}
    for uid, iid, true_r, est, _ in predictions:
        if uid not in top_n:
            top_n[uid] = []
        top_n[uid].append((iid, est))

    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n

# 상위 10개의 영화 추천
top_n = get_top_n_recommendations(predictions, n=10)

# 특정 사용자에 대한 상위 추천 영화 출력
print(top_n[str('힝이2')])
```
