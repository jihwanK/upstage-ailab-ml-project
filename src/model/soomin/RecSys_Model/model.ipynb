```python
import pandas as pd

# 평점에 따른 레이블 부여 (1~3점은 부정(0), 9~10점은 긍정(1))
def label_rating(rating):
    if rating >= 3:
        return 1  # 긍정
    elif rating < 3:
        return 0  # 부정
    else:
        return None  # 중립 혹은 제거 대상

data['label'] = data['review_point'].apply(label_rating)

# 중립(4~8점)의 데이터는 제거
data = data.dropna(subset=['label'])

print(data.head())  # 확인
```
