```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

user_olive_matrix = data.pivot_table(index='user_name', columns='product_name', values='review_point').fillna(0)

user_similarity = cosine_similarity(user_olive_matrix)

user_similarity_df = pd.DataFrame(user_similarity, index=user_olive_matrix.index, columns=user_olive_matrix.index)

# 유사도 확인
print("User Similarity Matrix:")
user_similarity_df
```

```python
def recommend_product(user_id, user_olive_matrix, user_similarity_df, n_recommendations=10):
    user_similarities = user_similarity_df[user_id]
    
    user_ratings = user_olive_matrix.loc[user_id]
    
    not_buy = user_ratings[user_ratings == 0].index
    
    weighted_ratings = user_olive_matrix[not_buy].T.dot(user_similarities)

    recommendations = weighted_ratings.sort_values(ascending=False).head(n_recommendations)
    
    return recommendations


user = '엠제쓰'
recommended_product = recommend_product(user, user_olive_matrix, user_similarity_df)
print(f'\n{user}을 위한 추천 목록!!!')
print(recommended_product)
```
