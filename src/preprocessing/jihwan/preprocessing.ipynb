```python
combined_df = pd.DataFrame()
for cat in os.listdir(os.path.join(data_dir, "reviews")):
    for file in os.listdir(os.path.join(data_dir, "reviews", cat)):
        tmp_df = pd.read_csv(os.path.join(data_dir, "reviews", cat, file))
        combined_df = pd.concat([combined_df, tmp_df[["brand_name", "product_name", "user_code", "review_rating"]]], axis=0, ignore_index=True)
combined_df
```
